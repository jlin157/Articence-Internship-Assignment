# Q1. suggest which products should be dropped from selling in the next year moving forward and which products should be sold more. Should any region be given preference over the other? hint: look at the profitability

```python
df = pd.read_csv(r'Data Science Evaluation.csv')
df
```

```python
df1=df[['Item Type','Fiscal Year','Profit as % of Cost']].sort_values(by=['Item Type','Fiscal Year'])
df1.reset_index(drop=True)
```

```python
products=set(df1['Item Type'])
products
```

```python
for i in products:
  print(i)
  print(df1[df1['Item Type']==i].groupby(['Fiscal Year']).describe())
```

## By looking at the Profitability, we can see that through out the 8 fiscal year. Cloths has always been the most profitable product while Meat is the least profitable one. Therefore, we should drop Meat from selling and sell more Cloths

```python
df1=df[['Region','Item Type','Profit as % of Cost']].sort_values(by=['Profit as % of Cost'])
df1=df1.reset_index(drop=True)
df1.groupby(['Region']).describe()
```

## Sub-Saharan Africa and Eruope regions should be given higher preference over the others, espeically Sub-Saharan Africa. Since two region contributed more than half of the total order count while Sub-Saharan Africa has the highest profitability on average.

# Q2. Which products are the easiest to sell. (hint: look at the ship date and order date) by region. Is there a relationship between products or region

```python
from datetime import datetime
```

```python
df['Ship Date']= pd.to_datetime(df['Ship Date'])
df['Order Date']= pd.to_datetime(df['Order Date'])
df
```

```python
df2=df[['Region','Item Type']]
df2['Days take to start deliver']=df['Ship Date']-df['Order Date']
df2
```

```python
regions=set(df2['Region'])
regions
```

```python
df2['Days take to start deliver']=(df2['Days take to start deliver'] / np.timedelta64(1, 'D')).astype(int)
```

```python
for i in regions:
  print('\n',i)
  print(df2[df2['Region']==i].groupby('Item Type').mean())
```

## In Sub-Saharan Africa, Meat is the easiest to sell with the least days take to start a delievery.
## In Asia, Beverages is the easiest to sell.
## In Europe, Snacks is the easiest to sell.
## In Central America and the Caribbean, Cosmetics is the easiest to sell.
## In Australia and Oceania, Fruits is the easiest to sell.
## In Middle East and North Africa, Office Supplies is the easiest to sell.
## In North America, Cereal is the easiest to sell.
